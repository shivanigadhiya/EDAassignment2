Discuss the advantages and disadvantages of different imputation techniques.

There is no missing value in the dataset. If it has then we can handle the missing values using mean , median and mode technique.

1.mean : If our data is likely normal distribution then we can apply mean technique. But if the dataset has outlier then we can not use mean because it can effect our model performance.

2.median : It is used to handle outliers.It does not take into account the precise value of each observation and hence does not use all information available in the data. 

3.mode : It can be located graphically, unlike mean and median. It can be used for qualitative analysis.The mode does not consider all the values in the data.
